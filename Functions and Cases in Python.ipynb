```python
#Importing Python Libraries required

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Reading dataset

df = pd.read_csv("youtube_dataset.csv")
```

```python
#Sorting dataframe in descending order of subscriber count

sorted_df=df.sort_values(by='subscribers', axis=0, ascending= False)
```

```python
#Filtering first 1000 rows and creating new dataframe

filtered_df = sorted_df.iloc[:1000,:]
```

```python
#Output of last 5 rows : Shows first 1000 rows have been filtered.

filtered_df.tail()
```

```python
#Defining function

def Distribution_of_Channel_Type_Count(col):
    
    y= (filtered_df[col]).value_counts()
    print(y)
```

```python
Distribution_of_Channel_Type_Count('channeltype')
```
